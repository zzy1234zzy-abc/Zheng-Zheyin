```python
function buyVotes() public payable notEnded {
    require(msg.value >= 1 ether);

    // Check: validate input
    uint256 voteCount = msg.value / 1e18;
    uint256 remainder = msg.value % 1e18;

    // Effect: update state first
    remainingVotes[msg.sender] += voteCount;

    // Interaction: send back change
    if (remainder > 0) {
        msg.sender.transfer(remainder);
    }
}
```
